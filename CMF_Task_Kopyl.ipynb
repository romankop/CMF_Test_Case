Моя стратегия предполагает парный трейдинг акций двух ETF, которые вкладываются в казначейские облигации США разной длительности.

- TLT: iShares 20+ Year Treasury Bond ETF (инвестиции в облигации с оставшимся сроком длительности от 20 лет)
- IEI: iShares 3-7 Year Treasury Bond ETF (инвестиции в облигации с оставшимся сроком длительности от 3 до 7 лет)

В качестве бенчмарка выбраны акции одного ETF казначейских облигаций с оставшимся сроком длительности от 1 года (GOVT). Все инструменты торгуются на бирже NASDAQ.

Стратегия будет реализовываться с помощью Фильтра Калмана. Стратегия работает на дневных данных.

Сначала проведем описательный анализ.

Посмотрим, как меняется оценка коэффициента хеджирования в течение обучающей выборки и сравним на цены двух активов.

```python
from __future__ import print_function

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas_datareader.yahoo.daily import YahooDailyReader
from pykalman import KalmanFilter

def draw_date_coloured_scatterplot(etfs, prices):

    plen = len(prices)
    colour_map = plt.cm.get_cmap('YlOrRd')    
    colours = np.linspace(0.1, 1, plen)
    
    scatterplot = plt.scatter(
        prices[etfs[0]], prices[etfs[1]], 
        s=30, c=colours, cmap=colour_map, 
        edgecolor='k', alpha=0.8
    )
    
    colourbar = plt.colorbar(scatterplot)
    colourbar.ax.set_yticklabels(
        [str(p.date()) for p in prices[::plen//9].index]
    )
    plt.xlabel(prices.columns[0])
    plt.ylabel(prices.columns[1])
    plt.show()


def calc_slope_intercept_kalman(etfs, prices):

    delta = 1e-5
    trans_cov = delta / (1 - delta) * np.eye(2)
    obs_mat = np.vstack(
        [prices[etfs[0]], np.ones(prices[etfs[0]].shape)]
    ).T[:, np.newaxis]
    
    kf = KalmanFilter(
        n_dim_obs=1, 
        n_dim_state=2,
        initial_state_mean=np.zeros(2),
        initial_state_covariance=np.ones((2, 2)),
        transition_matrices=np.eye(2),
        observation_matrices=obs_mat,
        observation_covariance=1.0,
        transition_covariance=trans_cov
    )
    
    state_means, state_covs = kf.filter(prices[etfs[1]].values)
    return state_means, state_covs    
    

def draw_slope_intercept_changes(prices, state_means):

    pd.DataFrame(
        dict(
            slope=state_means[:, 0], 
            intercept=state_means[:, 1]
        ), index=prices.index
    ).plot(subplots=True)
    plt.show()



etfs = ['TLT', 'IEI']
start_date = "2015-01-01"
end_date = "2019-12-31"    

prices = YahooDailyReader(etfs, start_date, end_date).read()['Adj Close']

draw_date_coloured_scatterplot(etfs, prices)
state_means, state_covs = calc_slope_intercept_kalman(etfs, prices)
draw_slope_intercept_changes(prices, state_means)
```

С обучающей части я выберу те модификации, которые попали либо в 7 лучших по Шарпу, либо по отношению дохода к просадке.

На тестовой выборке я выберу одну модификацию, которую буду использовать на финальном прогоне.

```python
from kalman_etf_backtest import backtest
from IPython.display import clear_output
import pandas as pd

train_sharpe = {} # sharp values
train_cum_dd = {} # cum return / max drawdown values

params = []

for returns in (True, False):
    for training in (True, False):
        for scale in (0.5, 1.0, 1.5, 2.0):
            if training:
                for train_years_num in range(1, 6):
                    for use_ols in (True, False):
                        params.append({'training': training,
                                       'train_years_num': train_years_num,
                                       'use_ols': use_ols,
                                       'scale': scale,
                                       'returns': returns})
            else:
                params.append({'training': training,
                               'train_years_num': False,
                               'use_ols': False,
                               'scale': scale,
                               'returns': returns})

for param in params:
    clear_output(wait=True)
    res = backtest(tickers="TLT,IEI", start_date='2015-01-01', end_date='2019-12-31',
                   **param)
    train_sharpe[tuple(param.values())] = res['sharpe']
    train_cum_dd[tuple(param.values())] = res['cum_ret/max_dd']
    
train_sharpe_best = set(pd.Series(train_sharpe).sort_values(ascending=False)[:7].index.tolist())
train_cum_dd_best = set(pd.Series(train_cum_dd).sort_values(ascending=False)[:7].index.tolist())
train_best = train_sharpe_best.union(train_cum_dd_best)
clear_output()
```

```python
test_sharpe = {} # sharp values
test_cum_dd = {} # cum return / max drawdown values

for params in train_best:
    res = backtest(tickers="TLT,IEI", start_date='2020-01-01', end_date='2020-12-31',
                   training=params[0], train_years_num=params[1], use_ols=params[2],
                   scale=params[3], returns=params[4])
    clear_output(wait=True)
    test_sharpe[params] = res['sharpe']
    test_cum_dd[params] = res['cum_ret/max_dd']
    
test_sharpe_best = set(pd.Series(test_sharpe).sort_values(ascending=False)[:2].index.tolist())
test_cum_dd_best = set(pd.Series(test_cum_dd).sort_values(ascending=False)[:2].index.tolist())
clear_output()
```

```python
print("Sharpe Test Results:")
print(pd.Series(test_sharpe).sort_values(ascending=False))

print("Cum Return / Max Drawdown:")
print(pd.Series(test_cum_dd).sort_values(ascending=False))
```

Лучший результат по обоим показателям показала стратегия с 4 годами обучения и масштабом 1. Такую модификацию я и буду использовать на финальной выборке.
